```python
import pandas as pd
import requests as rq
import threading as tr
```

```python
url ='https://api.irbroker.com/api/v1/customerCreditPurchase'
headers = {'x-client-token': '989c3c4a-95a9-4eae-a6e3-203f246b2c1c'}
GLOBAL_LIST = []
```

```python
def get_response(month:int):
    responses = []
    for i in range(1, 31+1):
        if i < 10 :
            i = '0' + str(i)
            
        if month < 10 :
            month = '0' + str(month)
        
            
        data = {
            'dsCode': 764,
            'lastRemainFrom': 100000,
            'size': -1,
            'dateTo': f'1403/{month}/{i}'
        }
        
        try:
            response = rq.get(url=url, headers=headers, params=data).json()['result']
            responses += response
            
        except Exception as e:
            print(f"Error fetching data for {month}/{i}: {e}")
            
    global GLOBAL_LIST
    GLOBAL_LIST += responses
    
    return responses
```

```python
thread_list = []

for m in range(1, 3+1):
    t = tr.Thread(target=get_response, args=(m,))
    thread_list.append(t)
    t.start()
    
for thread in thread_list:
    thread.join()
```
